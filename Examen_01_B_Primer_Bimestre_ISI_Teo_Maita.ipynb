```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
display(df.head(10))
display(df.describe())
```

### KPI 1: Tasa de Rotación de Personal

**Fórmula Matemática Utilizada:**

$$\text{Tasa de deserción} = \left( \frac{\text{Empleados que salieron (Attrition = Yes)}}{\text{Total de Empleados Evaluados}} \right) \times 100$$

**Valores Numéricos Obtenidos:**
  * **Total de Empleados Evaluados:** 1470
  * **Empleados que salieron (Yes):** 237
  * **Resultado del KPI:** 16.12%
  * **Período de tiempo:** Ciclode tiempo operativo en curso

**Interpretación del negocio:**
Una tasa de rotación del 16.12% indica un nivel bajo-moderado de deserción que, aunque común en la industria, requiere monitoreo preventivo continuo. Es crucial que el Sistema de Planificación de Recursos Empresariales (ERP) unifique estos datos con encuestas de entorno laboral para identificar patrones predictivos antes de que impacten los costos de reclutamiento.

```python
total_empleados = len(df)
empleados_salieron = len(df[df['Attrition'] == 'Yes'])
attrition_rate = (empleados_salieron / total_empleados) * 100
no_salieron = total_empleados - empleados_salieron
print("KPI 1: TASA DE DESERCIÓN LABORAL")
print(f"Total de Empleados Evaluados: {total_empleados}")
print(f"Empleados que desertaron: {empleados_salieron}")
print(f"Empleados que permanecieron: {no_salieron}")
print(f"Porcentaje de decersión laboral: {attrition_rate:.2f}%")
```

```python
plt.figure(figsize=(7, 7))
plt.pie([empleados_salieron, no_salieron],
        labels=['Empleados que desertaron', 'Empleados que permanecen'],
        autopct='%1.2f%%',
        startangle=90,
        colors=['Red', 'Blue'])
plt.axis('equal')
plt.title('KPI 1: Distribución Rotación de Personal', fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```

###KPI 2: Ingreso Promedio Mensual por Departamento

* **Fórmula Matemática Utilizada:**
$$\text{Ingreso Promedio} = \frac{\text{Suma de Ingresos del Departamento}}{\text{Número de Empleados del Departamento}}$$

* **Valores Numéricos Obtenidos:**
  - **Ingreso Promedio en Ventas:** 6,959.17 USD
  - **Ingreso Promedio en Investigación y Desarrollo:** 6,281.25 USD
  - **Ingreso Promedio en Recursos Humanos:** 6,654.51 USD
  - **Brecha Ventas vs. Investigación y Desarrollo:** 677.92 USD
  - **Brecha Ventas vs. Recursos Humanos:** 304.66 USD

* **Interpretación de Negocio:**
Se revela una asimetría salarial donde el área de ventas lidera los ingresos promedio que supera con cientos de dólares los otros departamentos. Desde la perspectiva de la gobernanza corporativa se debe evaluar si esta diferencia responde a incentivos variables por cumplimiento de cuotas o si constituye una inequidad estructural.

```python
df_ventas = df[df['Department'] == 'Sales']
df_id = df[df['Department'] == 'Research & Development']
df_rh = df[df['Department'] == 'Human Resources']
promedio_ventas = df_ventas['MonthlyIncome'].mean()
promedio_id = df_id['MonthlyIncome'].mean()
promedio_rh = df_rh['MonthlyIncome'].mean()
print("KPI 2: AUDITORÍA DE COMPENSACIÓN")
print(f"Ingreso Promedio en ventas: ${promedio_ventas:.2f}")
print(f"Ingreso Promedio en Investigación y desarrollo: ${promedio_id:.2f}")
print(f"Ingreso Promedio en Recursos Humanos (RRHH): ${promedio_rh:.2f}")
print("\n")
print("ANÁLISIS DE BRECHAS SALARIALES ESTADÍSTICAS:")
print(f"Brecha ventas vs. Investigación y desarrollo: ${(promedio_ventas - promedio_id):.2f}")
print(f"Brecha ventas vs. Recursos humanos: ${(promedio_ventas - promedio_rh):.2f}")
```

```python
departamentos = ['Ventas', 'Investigación y desarrollo', 'Recursos Humanos']
promedios = [promedio_ventas, promedio_id, promedio_rh]
colores = ['Blue', 'Red', 'Green']

plt.figure(figsize=(8, 5))
plt.bar(departamentos, promedios, color=colores, width=0.4)

plt.title('KPI 2: Auditoría de Estructuras de Compensación por Área Funcional', fontsize=13, pad=15)
plt.ylabel('Ingreso Promedio Mensual ($)', fontsize=11)
plt.xlabel('Departamentos Registrados', fontsize=11)
plt.ylim(0, max(promedios) * 1.15)


for i, v in enumerate(promedios):
    plt.text(i, v + 150, f"${v:.2f}", ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 3. KPI 3: Índice de Balance Vida-Trabajo por Cargo

* **Fórmula Matemática Utilizada:**
$$\text{Balance Promedio} = \frac{\text{Suma de Calificaciones de Balance del Cargo}}{\text{Total de Empleados en ese Cargo}}$$

* **Valores Numéricos Obtenidos**
  - El sistema de información analiza la distribución de los 1470 colaboradores cruzando la métrica de entorno frente a los roles operativos del ERP.
  - Los cargos con menor balance percibido corresponden a las líneas de supervisión y gestión operativa (como Managers), mientras que las áreas técnicas e intermedias muestran una estabilidad relativa cercana a los 3.00 puntos (Nivel Bueno).

* **Interpretación y Análisis de Gobernanza:**
Se evalúa la calidad del entorno laboral y actúa como un predictor temprano de desgaste del personal (burnout). Al segmentar la métrica por cada rol operativo (JobRole), la gerencia puede identificar con precisión qué puestos específicos sufren de sobrecarga de tareas o extensiones críticas de jornada.


```python
balance_por_cargo = df.groupby('JobRole')['WorkLifeBalance'].mean().sort_values(ascending=True)
print("KPI 3: ÍNDICE DE BALANCE VIDA-TRABAJO PROMEDIO POR CARGO")
print("\n")
print("Puntuaciones promedio por rol operativo (Escala 1 al 4):")
for cargo, valor in balance_por_cargo.items():
    print(f"Cargo: {cargo:<30} | Balance Promedio: {valor:.2f}")
```

```python
cargos = balance_por_cargo.index
promedios_balance = balance_por_cargo.values
plt.figure(figsize=(10, 5))
colores_balance = ['Blue']
bars = plt.barh(cargos, promedios_balance, color=colores_balance, height=0.5)
plt.title('KPI 3: Balance Vida-Trabajo Promedio por Cargo', fontsize=14, pad=20)
plt.xlabel('Índice de Balance (1: Malo, 2: Regular, 3: Bueno, 4: Excelente)', fontsize=11)
plt.ylabel('Cargos Registrados (JobRole)', fontsize=11)
plt.xlim(0, 4.5)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.05, bar.get_y() + bar.get_height()/2, f"{width:.2f}",
             va='center', ha='left', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 4. KPI 4: Brecha de Estancamiento Laboral por Departamento

**Fórmula Matemática Utilizada:**
$$\text{Brecha de Estancamiento} = \text{Promedio de Años en Puesto Actual} - \text{Promedio de Años desde Última Promoción}$$

**Valores Numéricos Obtenidos:**
  - En todos los departamentos evaluados en el sistema, el promedio de años que un colaborador pasa en su mismo cargo supera el tiempo transcurrido desde su última promoción.
  - El análisis registra brechas netas positivas de estancamiento en las tres áreas funcionales, lo que indica que las líneas de carrera internas presentan una ralentización estructural.

**Interpretación de Negocio:**
Este indicador mide la fluidez de las líneas de carrera internas. Una brecha positiva (más años en el puesto que desde el último ascenso) revela estancamiento laboral, lo que reduce la productividad, afecta la motivación e incrementa el riesgo de fuga de talento experimentado.

```python
df_ventas = df[df['Department'] == 'Sales']
df_id = df[df['Department'] == 'Research & Development']
df_rh = df[df['Department'] == 'Human Resources']

puesto_ventas = df_ventas['YearsInCurrentRole'].mean()
promo_ventas = df_ventas['YearsSinceLastPromotion'].mean()
brecha_ventas = puesto_ventas - promo_ventas

puesto_id = df_id['YearsInCurrentRole'].mean()
promo_id = df_id['YearsSinceLastPromotion'].mean()
brecha_id = puesto_id - promo_id

puesto_rh = df_rh['YearsInCurrentRole'].mean()
promo_rh = df_rh['YearsSinceLastPromotion'].mean()
brecha_rh = puesto_rh - promo_rh

print("KPI 4: BRECHA DE ESTANCAMIENTO LABORAL")
print(f"VENTAS:")
print(f"  - Promedio de años en puesto actual: {puesto_ventas:.2f} años")
print(f"  - Promedio de años desde última promoción: {promo_ventas:.2f} años")
print(f"  - Brecha neta de estancamiento: {brecha_ventas:.2f} años")
print(f"INVESTIGACIÓN Y DESARROLLO:")
print(f"  - Promedio de años en puesto actual: {puesto_id:.2f} años")
print(f"  - Promedio de años desde última promoción: {promo_id:.2f} años")
print(f"  - Brecha neta de estancamiento: {brecha_id:.2f} años")
print(f"RECURSOS HUMANOS:")
print(f"  - Promedio de años en puesto actual: {puesto_rh:.2f} años")
print(f"  - Promedio de años desde última promoción: {promo_rh:.2f} años")
print(f"  - Brecha neta de estancamiento: {brecha_rh:.2f} años")
```

```python
departamentos = ['Ventas', 'Investigación y desarrollo', 'Recursos Humanos']
promedios_puesto = [puesto_ventas, puesto_id, puesto_rh]
promedios_promo = [promo_ventas, promo_id, promo_rh]

x = np.arange(len(departamentos))
width = 0.35
plt.figure(figsize=(9, 5))

rects1 = plt.bar(x - width/2, promedios_puesto, width, label='Años en Puesto Actual', color='Blue')
rects2 = plt.bar(x + width/2, promedios_promo, width, label='Años desde Última Promoción', color='Red')

plt.title('KPI 4: Comparativa de Estancamiento y Líneas de Carrera por Área', fontsize=13, pad=15)
plt.ylabel('Promedio de Años', fontsize=11)
plt.xlabel('Departamentos Registrados', fontsize=11)
plt.xticks(x, departamentos)
plt.ylim(0, max(promedios_puesto) * 1.3)
plt.legend(loc='upper right')

def etiquetas_tiempo(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., height + 0.1, f"{height:.2f}",
                 ha='center', va='bottom', fontsize=10, fontweight='bold')

etiquetas_tiempo(rects1)
etiquetas_tiempo(rects2)

plt.tight_layout()
plt.show()
```

```python
departamentos = ['Ventas', 'Investigación y desarrollo', 'Recursos Humanos']
brechas_estancamiento = [brecha_ventas, brecha_id, brecha_rh]
colores = ['Blue', 'Red', 'Green']
plt.figure(figsize=(8, 5))
bars = plt.bar(departamentos, brechas_estancamiento, color=colores, width=0.4)

plt.title('KPI 4: Promedio de Años de Estancamiento Laboral por Departamento', fontsize=13, pad=15)
plt.ylabel('Años de Estancamiento Netos', fontsize=11)
plt.xlabel('Departamentos Registrados', fontsize=11)
plt.ylim(0, max(brechas_estancamiento) * 1.25)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.08, f"{height:.2f} años",
             ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
